```python
import pandas as pd

df = pd.read_csv("dataset/crime_data.csv.csv", low_memory=False)

print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

df.head()
```

```python
print(df.columns.tolist())
```

```python
import matplotlib.pyplot as plt

top_crimes = df['Primary Type'].value_counts().head(10)

print(top_crimes)

top_crimes.plot(kind='bar')

plt.title('Top 10 Crime Types')
plt.xlabel('Crime Type')
plt.ylabel('Number of Cases')

plt.tight_layout()
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day
df['Hour'] = df['Date'].dt.hour
df['Weekday'] = df['Date'].dt.day_name()

print(df[['Date','Month','Day','Hour','Weekday']].head())
```

```python
monthly_crime = df['Month'].value_counts().sort_index()

monthly_crime.plot(kind='bar')

plt.title('Crime Count by Month')
plt.xlabel('Month')
plt.ylabel('Number of Crimes')

plt.show()
```

```python
hourly_crime = df['Hour'].value_counts().sort_index()

hourly_crime.plot(kind='line')

plt.title('Crime Count by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Crimes')

plt.grid(True)

plt.show()
```

```python
top_districts = df['District'].value_counts().head(10)

print(top_districts)

top_districts.plot(kind='bar')

plt.title('Top 10 Crime Hotspots')
plt.xlabel('District')
plt.ylabel('Crime Count')

plt.show()
```
